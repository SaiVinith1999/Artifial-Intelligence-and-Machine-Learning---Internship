```python
df1=df.iloc[:,:].apply(lambda x:x.fillna(0))
df1.isnull().sum()
```

```python
df1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df1.Gender=le.fit_transform(df1.Gender)
df1.head()
```

```python
df1.Age=le.fit_transform(df1.Age)
df1.head()
```

```python
df1.City_Category=le.fit_transform(df1.City_Category)
df1.head()
```

```python
df1=df1.drop(['Stay_In_Current_City_Years','User_ID','Product_ID'],axis=1)
df1.head()
```

```python
import statsmodels.formula.api as smf
import numpy as np
df2=smf.ols("Purchase ~ Gender + Age + Occupation + City_Category+Marital_Status+Product_Category_1+Product_Category_2+Product_Category_3",data=df).fit()
df2.summary()
```

```python
exp=smf.ols("np.log(Purchase) ~ Gender + Age + Occupation + City_Category+Marital_Status+Product_Category_1+Product_Category_2+Product_Category_3",data=df).fit()
exp.summary()
```

```python
sqrt=smf.ols("np.sqrt(Purchase) ~ Gender +Age + Occupation + City_Category+Marital_Status+np.sqrt(Product_Category_1)+np.sqrt(Product_Category_2)+np.sqrt(Product_Category_3)",data=df).fit()
sqrt.summary()
```

```python
sqrt=smf.ols("(Purchase)**2 ~ Gender +Age + Occupation + City_Category+Marital_Status+(Product_Category_1)**(1/4)+(Product_Category_2)**(1/4)+(Product_Category_3)**(1/4)",data=df).fit()
sqrt.summary()
```
