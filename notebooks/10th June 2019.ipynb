```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn 
```

```python
df=pd.read_csv('Datafiles/area.csv')
df
```

```python
seaborn.pairplot(df.iloc[:,1:])
```

```python
import statsmodels.formula.api as smf
a=smf.ols("prices  ~ area",data=df).fit()
a.summary()
```

```python
loga=smf.ols("prices ~ np.log(area)",data=df).fit()
loga.summary()
```

```python
expa=smf.ols("np.log(prices)~area",data=df).fit()
expa.summary()
```

```python
sqrta=smf.ols("prices ~ np.sqrt(area)",data=df).fit()
sqrta.summary()
```

```python
sqrta=smf.ols("np.sqrt(prices)~area",data=df).fit()
sqrta.summary()
```

```python
sqrta=smf.ols("np.sqrt(prices)~np.sqrt(area)",data=df).fit()
sqrta.summary()
```

```python
x=df.iloc[:,1:-1]
y=df.iloc[:,2]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
y_pred
```

```python
y_test==y_pred
```

```python
print(np.mean(y_test))
print(np.mean(y_pred))
```

```python
plt.scatter(x_train,y_train,color='green')
plt.plot(x_train,reg.predict(x_train))
plt.title("Homeprices")
plt.xlabel("area")
plt.ylabel("prices")
```
